```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import collections
import os
import tensorflow as tf
from datetime import datetime
import csv
import numpy as np

f_array = open('x_in.npy', 'a+b')


tests = [
    # (Type, Test)
    (int, int),
    (float, float),
    (datetime, lambda value: datetime.strptime(value, "%Y/%m/%d"))
]

def getType(value):
     for typ, test in tests:
         try:
             test(value)
             return typ
         except ValueError:
             continue
     # No match
     return str

with open('x_in.csv', 'rt', encoding='mac_roman', newline='') as f:
    reader = csv.reader(f)
    #input_x = reader
    x_input = list(reader)

main_array = np.zeros(shape=(104000,22))
cnt_0 = 0
for request in x_input:
    #print("Counter: %s" % cnt_0)
    c_request = []
    c_request_list = []
    for element in request:
        element_type = getType(element)
        #print("Element Type: %s for %s" % (element_type,element))
        if element_type==int:
            c_request.append(int(element))
            #c_request_list.append(int(element))
        elif element_type==str:
            element_ascii = int(''.join(str(ord(c)) for c in element))
            c_request.append(element_ascii)
            #c_request_list.append(str(element))
        elif element_type==float:
            c_request.append(float(element))
            #c_request_list.append(int(element))
        else:
            print("Error")

    c_request = np.array(c_request)
    #print(c_request_list)
    #print(c_request)
    main_array[cnt_0] = c_request
    cnt_0 = cnt_0 + 1

np.save(f_array, main_array)
f_array.close()    

print("Loading:")
fi=open('x_in.npy','rb')
main_array_load = np.load(fi)

print(main_array_load.shape)
fi.close() 

#Creating Y_in
main_array_y = np.ones(shape=(104000,1))
print(main_array_y.shape)
print(main_array_y[0:10])
fy_array = open('y_in.npy', 'a+b')
np.save(fy_array, main_array_y)
fy_array.close()    
```
